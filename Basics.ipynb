```python
import pygame,sys
from pygame.locals import *

pygame.init()

FPS = 30
clock = pygame.time.Clock()
window_l = 500
window_w = 300
DISPLAYSURF = pygame.display.set_mode((window_l,window_w),0,32)
pygame.display.set_caption('Simple game go to rect and press space')

BLACK = (0,0,0)
WHITE = (255,255,255)
#img = pygame.image.load('Art//idle_r.png')
img_r = pygame.image.load('Art//idle_r.png')
img_l = pygame.image.load('Art//idle_l.png')
w_l = pygame.image.load('Art//walk_l.png')
w_r = pygame.image.load('Art//walk_r.png')
img = img_r
xpos = window_l/2
ypos = window_w-57


l_att = pygame.image.load('Art//attack_left.png')
r_att = pygame.image.load('Art//attack_right.png')
x=0
y=0
last_right = True
last_left = False
Flag=0
def attack(x,y,xpos,ypos,last_right,last_left,rect):
    if last_right == True:
        #load the right attack png
        print('right facing')
        dr = DISPLAYSURF.blit(img,(xpos,ypos))
        print(dr==None)
        if dr.right >= rect.left:
            print("contactr")
            x=1000
            y=1000
    elif last_left == True:
        #load the left attack png
        print('left facing')
        dr = DISPLAYSURF.blit(l_att,(xpos,ypos))
        if dr.left <= rect.right:
            print("contactl")
            x=1000
            y=1000
        return x,y
    
    

while True:
    DISPLAYSURF.fill(WHITE)
    rect = pygame.draw.rect(DISPLAYSURF,BLACK,((window_l/2)+100+x,window_w-58+y,100,50))
    DISPLAYSURF.blit(img,(xpos,ypos))
#    DISPLAYSURF.blit(img,(catx,caty))
    
    for event in pygame.event.get():
#       print(event)
        if event.type == KEYDOWN:
            if event.key == K_UP:
                ypos = ypos-5
        if event.type == KEYDOWN:
            if event.key == K_DOWN:
                ypos = ypos+5
        if event.type == KEYDOWN:
            if event.key == K_LEFT:
                last_left = True
                last_right = False
                xpos = xpos-5
                if Flag == 0:
                    img = w_l
                    Flag =1
                    print(Flag)
                    continue
                if Flag == 1:
                    img = img_l
                    Flag=0
                    continue
                print(Flag)
        if event.type == KEYDOWN:
            if event.key == K_RIGHT:
                last_left = False
                last_right = True
                xpos = xpos+5
                if Flag == 0:
                    img = w_r
                    Flag=1
                    print(Flag)
                    continue
                if Flag == 1:
                    img = img_r
                    Flag=0
                    continue
                print(Flag)
        if event.type == KEYDOWN:
            if event.key == K_SPACE:
                print('space pressed')
                print(xpos==None)
                print(ypos==None)
                print(x==None)
                print(y==None)
                print(last_left==None)
                print(last_right==None)
                x,y = attack(x,y,xpos,ypos,last_right,last_left,rect)
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
        #pygame.draw.rect(DISPLAYSURF,BLACK,(xpos,ypos,100,50))
        
        #load the normal png
            
        
        pygame.display.update()
        clock.tick(FPS)
```
